# ANALYSE ET TRANSFORMATION DU DATA SET traindemographics.csv
Le but de cette partie consiste à explorer et comprendre le data set **traindemographics.csv**. Puis faire les transformations neccesaire à contruction d'un modlèle robuste et efficace. les autres data set seront ensuite traités dans d'autres parties.

## Interpretation

### Type des colonnes
2 colonnes de type float (longitude_gps, latitude_gps) et 7 colonnes de type object (customerid, birthdate, bank_branch_client, bank_account_type, employment_status_clients et level_of_education_clients)

```python
#inspection de colonnes contenant des valeurs nulls
df_traindemographics.isna().sum()
```

```python
#pourcentage de valeur null dans chaque colonne
(df_traindemographics.isna().sum()/df_traindemographics.shape[0])*100
```

## Interpretation
il y a plus 98% de valeurs non renseignées dans la colonne **bank_branch_clients**, plus de 86% dans la colonne **level_of_education_clients**. Pour la suite de notre analyse ces colonnes seront tout simplement supprimées de notre data set. la colonne **employment_status_clients** quant à elle contient moins de 15% de valeurs nulls. Nous essayerons par la suite traiter ces valeurs manquantes. Cette colonne semble importante car l'emploi occuppé par un individu influence enormement la capacité à rembourser ou non un prêt.

```python
#Traitement de la colonne Birthday
#transformons le type cette colonne en Date
df_traindemographics['birthdate'] = pd.to_datetime(df_traindemographics['birthdate'])

#création de la colonne Age à partir de la colonne birthdate
#créaton de la colonne birthyear
df_traindemographics['birthyear'] = df_traindemographics['birthdate'].dt.year

#Création de la fonction de calcul age
def calAge(x):
    return (2022 - x)

#Création de la colonne age
df_traindemographics['age'] = df_traindemographics['birthyear'].apply(calAge)

#Suppression des colonnes qui ne serviront pas dans la suite de notre analyse
df_traindemographics = df_traindemographics.drop(['birthdate', 'bank_branch_clients', 'level_of_education_clients', 'birthyear'], axis = 1)
df_traindemographics.head()
```

## Interpretation
à ce stade de notre analyse nous avons supprimer les colonnes **bank_branch_clients** et  **level_of_education_clients** Car elles contenaient chacune plus de 80% de valeurs manquantes. Nous avons aussi transformé la colonne **birthdate** en *age* car il plus sera plus bénéfique pour nous de travailler avec l'age d'un individus plutôt que sa date de naissance dans la suite de notre analyse.

```python
#inspection de la colonne *employment_status_clients*
df_traindemographics['employment_status_clients'].value_counts(normalize=True)*100
```

Au vue de la représentation des différentes catégories présentes dans la colonne **employment_status_clients**, il est très difficille d'affecter selon un principe non aléatoire des valeurs aux données manquantes. L'idéal serait de s'addresser au fournisseur de données pour comprendre comment gérer les valeurs manquantes. A défaut de la présence du fournisseur de donnée, nous décisidons finalemment de supprimer les individus ayant des valeurs nulls pour cette colonne. ## Interpretation 
Nos colonnes **closeddate** ,**firstduedate** et **firstrepaiddate** sont désormais de type **datetime64**, on peut donc mettre en place de critères de detection de statut des clients
statut = 1 ==> bon client, statut = 0 ==> mauvais client.

```python
#si pour un client donnée, l'emsemble de ses prêts respectent la règle :
#termdays <= 30 et firstrepaiddate <= closeddate alors c'est un bon client (statut = 1 , => suceptible de rembourser)
#si non client douteux (statut = 0, => non suceptible de rembouser)
```

```python
df_trainprevloans.info()
```

```python
#Transformation des colonnes closeddate et firstrepaiddate

#closeddate
df_trainprevloans['year'] = df_trainprevloans['closeddate'].dt.year
df_trainprevloans['month'] = df_trainprevloans['closeddate'].dt.month
df_trainprevloans['day'] = df_trainprevloans['closeddate'].dt.day
df_trainprevloans['closeddate'] = pd.to_datetime(df_trainprevloans[['year', 'month', 'day']], errors = 'coerce')
df_trainprevloans = df_trainprevloans.drop(['year', 'month', 'day'], axis = 1)

#firstrepaiddate
df_trainprevloans['year'] = df_trainprevloans['firstrepaiddate'].dt.year
df_trainprevloans['month'] = df_trainprevloans['firstrepaiddate'].dt.month
df_trainprevloans['day'] = df_trainprevloans['firstrepaiddate'].dt.day
df_trainprevloans['firstrepaiddate'] = pd.to_datetime(df_trainprevloans[['year', 'month', 'day']], errors = 'coerce')
df_trainprevloans = df_trainprevloans.drop(['year', 'month', 'day'], axis = 1)

df_trainprevloans
```

```python
df_trainprevloans.dtypes
```

```python
df_trainprevloans
```

```python
#indification des bons et mauvais client.
# si date de remboursement ou 1ère versement respectée alors bon client(statut = 1) sinon mauvais client (statut = 0)
import numpy as np
df_trainprevloans['statut'] = np.where(
    (df_trainprevloans['firstrepaiddate'] > df_trainprevloans['firstduedate']),
    0 ,
    1
)
```

```python
#sauvegarder du data set pour prochaine analyse 
df_trainprevloans.to_csv("C:/Users/menan/Data_science/Projet/20220324 - Loan_Default_Prediction/2_PreparedData/bestTrainprevloans.csv")
```

```python
dictionnaire_fichier = {}
for i in range(len(df_trainprevloans['customerid'])):
    a = df_trainprevloans['customerid'][i]
    if a not in dictionnaire_fichier.keys():
        my_list = []
        my_list.append(df_trainprevloans['statut_prevloans'][i])
        dictionnaire_fichier[a] = my_list
    else:
        dictionnaire_fichier[a].append(df_trainprevloans['statut_prevloans'][i])
```

```python
#en partant du dico initialement créé, on crée les varaibles customerId, et statut qui contiendrons les valeurs de notre good_data_set
liste_keys = list(dictionnaire_fichier.keys())
customerId = [] 
statut = []
for i in range(len(liste_keys)):
    customerId.append(liste_keys[i])
    a = sum(dictionnaire_fichier[liste_keys[i]])
    b = len(dictionnaire_fichier[liste_keys[i]])
    statut.append((a/b))
```

```python
good_data_set = pd.DataFrame({'customerId' : customerId, 'statut' : statut})
good_data_set
```
